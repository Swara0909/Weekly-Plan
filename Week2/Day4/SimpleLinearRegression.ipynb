```python
# Simple Linear Regression
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
#Load the dataset
data=pd.read_csv(r"C:\Users\SwaraNafade\Weekly-Plan\Week2\Day4\data.csv")
```

```python
#Variables
x=data[["ExperienceYears"]] #Input or independent variable
y=data["Salary"] #dependent variable

#Creating a linear regression model
model=LinearRegression()

#Train the model
model.fit(x,y)

#predict y based on data
output=model.predict(x)

#coefficients
print("Slope:",model.coef_[0])
print("Intercept:",model.intercept_)
```

```python
# Plot the output
plt.scatter(x, y, color="blue", label="Actual Data")
plt.plot(x, output, color="red", label="Regression Line")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Linear Regression Example")
plt.legend() #shows an box
'''Blue dots → Actual Data
Red line → Regression Line'''
plt.show()
```
